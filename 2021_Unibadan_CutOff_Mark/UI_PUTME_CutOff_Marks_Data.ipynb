### Importing libraries

```python
import pandas as pd
from bs4 import BeautifulSoup
from csv import writer
import requests
```

### Scraping the data

```python
url = 'https://www.ui.edu.ng/news/20212022-post-utme-cut-marks'
webpage = requests.get(url)
webpage
```

### Accessing the content of the webpage

```python
webpage_content = BeautifulSoup(webpage.content, 'html.parser')
webpage_content
```

### Accessing the Table in the webpage

```python
table = webpage_content.find('table')
table
```

### Accessing the rows in the table

```python
table_rows = table.find_all('tr')
table_rows
```

### Accessing the data in each rows

```python
for row in table_rows[1:]:
    
    row_data = row.find_all('td')
    
    clean_data = [values.get_text(strip = True) for values in row_data]
    print(clean_data)
```

### Creating a csv file for the data

```python
with open('2021_Unibadan_Cutoffmark.csv','w', encoding = 'utf-8', newline = '') as file:
    
    filewriter = writer(file)
    
    header = ['PROGRAMME', 'MERIT', 'CATCHMENT', 'ELDS']
    filewriter.writerow(header)
    
    for row in table_rows[1:]:
    
        row_data = row.find_all('td')

        clean_data = [values.get_text(strip = True) for values in row_data]
        filewriter.writerow(clean_data)
```
