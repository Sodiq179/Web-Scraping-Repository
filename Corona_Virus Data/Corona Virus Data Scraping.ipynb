### Importing libraries

```python
from bs4 import BeautifulSoup
import requests
from csv import writer
import pandas as pd
```

### Scraping the data

```python
site_url = 'https://covid19.ncdc.gov.ng/'
site_page = requests.get(site_url)
site_page
```

```python
page_content = BeautifulSoup(site_page.content, 'html.parser')
page_content
```

```python
table = page_content.find('table')
table
```

```python
table_rows = table.find_all('tr')
table_rows
```

```python
for row in table_rows[1:]:
    row_data = row.find_all('td')
    
    row_elements = [values.get_text(strip = True) for values in row_data]
    print(row_elements)
```

### Saving the dataset as a csv file

```python
with open('Corona_Virus_Data.csv','w', encoding = 'utf-8', newline = '') as file:
    filewriter = writer(file)
    
    header = ['States Affected','No. of Cases (Lab Confirmed)','No. of Cases (on admission)','No. Discharged','No. of Deaths']
    filewriter.writerow(header)
    
    for row in table_rows[1:]:
        row_data = row.find_all('td')

        row_elements = [values.get_text(strip = True) for values in row_data]
        filewriter.writerow(row_elements)
```
