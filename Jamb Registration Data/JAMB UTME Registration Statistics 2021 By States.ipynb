### Importing Necessary Libraries

```python
import pandas as pd
from bs4 import BeautifulSoup
from csv import writer
import requests
```

### Scraping the Dataset

```python
url = 'https://newngtimes.com/jamb-registration-statistics/'
webpage = requests.get(url)
webpage
```

```python
webpage_content = BeautifulSoup(webpage.content, 'html.parser')
webpage_content
```

```python
table = webpage_content.find('table')
table
```

```python
table_rows = table.find_all('tr')
table_rows
```

```python
for row in table_rows[1:]:
    row_data = row.find_all('td')
    
    data_cleaned = [values.get_text(strip = True) for values in row_data]
    print(data_cleaned)
```

```python
with open('Jamb_Registration.csv','w',encoding = 'utf-8', newline = '') as file:
    filewriter = writer(file)
        
    header = ['S/N', 'STATE', 'TOTAL', 'PERCENTAGE(%)']
    filewriter.writerow(header)
    
    for row in table_rows[1:]:
        row_data = row.find_all('td')

        data_cleaned = [values.get_text(strip = True) for values in row_data]
        filewriter.writerow(data_cleaned)
```
